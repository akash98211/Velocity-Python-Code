# strftime

```python
todays_date = datetime.datetime.now()
print("Todays date is :",todays_date)
```

```python
todays_date.strftime('%Y')  # Year 2021
todays_date.strftime('%y')  # Year 21
todays_date.strftime('%d')  # Date
todays_date.strftime('%m')  # Month
todays_date.strftime('%b')  # Abbreviated Month name 
todays_date.strftime('%B')  # Full Month name 
todays_date.strftime('%D')  # Date 
todays_date.strftime('%a')  # Abbreviated weekday
todays_date.strftime('%A')  # weekday
todays_date.strftime('%w')  # weekday Number Sun = 0, Mon = 1, Tue = 2, Wed = 3
todays_date.strftime('%j')  # Day Number in Year >> 286
todays_date.strftime('%p')  # AM/PM
todays_date.strftime('%U')  # week Number in Year
```

```python
# 13/10/2021

todays_date.strftime('%d/%m/%Y')
```

```python
todays_date.strftime('%d-%m-%Y')
```

```python
# 13 Oct 2021
todays_date.strftime('%d %b %Y')
```

# strptime 

```python
String points to time
Create an datetime object from given string
```

```python
string = '20 Aug 2021'

d1 = datetime.datetime.strptime(string, '%d %b %Y' )
print(d1.date())
```

```python
string = '20 Aug 2021'
d1 = datetime.datetime.strptime(string, '%d %b %Y' )
d1.strftime('%D')
```

```python
date_list = ['13/10/2021', '13 Oct 2021', '13-10-2021','15/10/2021', '13-10-21', '13/10/21',
            '2021 Oct 13', '2021/10/13','15 Oct 2021','10/13/2021','20/10/2021']

date_format_list = ["%d/%m/%Y","%d/%m/%y","%d %b %Y", '%d-%m-%Y','%d-%m-%y','%Y %b %d',
              "%Y/%m/%d",'%m/%d/%Y']

def get_date(input_date):
    for date_format in date_format_list:
        try:
            d1 = datetime.datetime.strptime(input_date,date_format)
            return d1
        except:
            continue
    
new_date= get_date('2021/10/13')
print(new_date.date())

```
