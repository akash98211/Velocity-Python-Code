# 13. symmetric_difference

```python
set1 = {'Pune','Mumbai','Delhi','Chennai'}
set2 = {'Pune','Kolkata','Bangalore','Delhi'}

set3 = set1.symmetric_difference(set2)
print(set1,len(set1))  #  {'Pune','Mumbai','Delhi','Chennai'}
print(set2,len(set2))  # {'Pune','Kolkata','Bangalore','Delhi'}
print(set3,len(set3)) # Mumbai, Chennai,Kolkata,Bangalore
```

```python
set1 = {2,3,5,10,20}
set2 = {10,20,6,9,8}

set3 = set1.symmetric_difference(set2)
print(set1,len(set1))
print(set2,len(set2))
print(set3,len(set3)) 
```
